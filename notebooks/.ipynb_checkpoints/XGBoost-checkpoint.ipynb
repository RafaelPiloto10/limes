```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics  
from selection import get_dataset
xTrain, xTest, yTrain, yTest = get_dataset("../dataset/training.csv", True)
```

```python
clf = XGBClassifier(n_estimators = 100)  

param_grid = { 
    'n_estimators': [100, 500, 1000],
    'max_depth' : [5, 10, 500, 1000],
}
gscv = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, verbose=2)
gscv.fit(xTrain, yTrain["IsBadBuy"].ravel())
gscv.best_params_
```

```python
# {'criterion': 'gini',
#  'max_depth': 8,
#  'max_features': 'sqrt',
#  'n_estimators': 200}
clf = XGBClassifier(n_estimators = 100, max_depth=4)  
clf.fit(xTrain, yTrain["IsBadBuy"].ravel())
y_pred = clf.predict(xTest)
metrics.accuracy_score(yTest["IsBadBuy"].ravel(), y_pred)
```
