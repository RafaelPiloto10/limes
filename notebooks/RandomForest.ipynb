```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics  
from selection import get_balanced_dataset
xTrain, xTest, yTrain, yTest = get_balanced_dataset("../dataset/training.csv", True)
```

```python
clf = RandomForestClassifier(n_estimators = 100)  

param_grid = { 
    'n_estimators': range(100, 1100, 100),
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
gscv = GridSearchCV(estimator=clf, param_grid=param_grid, cv= 5)
gscv.fit(xTrain, yTrain["IsBadBuy"].ravel())
gscv.best_params_
```

```python
# {'criterion': 'gini',
#  'max_depth': 8,
#  'max_features': 'auto',
#  'n_estimators': 100}
clf = RandomForestClassifier(n_estimators = 200, max_depth=15, criterion="gini")  
clf.fit(xTrain, yTrain["IsBadBuy"].ravel())
y_pred = clf.predict(xTest)
metrics.accuracy_score(yTest["IsBadBuy"].ravel(), y_pred)
```

```python
from sklearn.metrics import roc_auc_score
clf = RandomForestClassifier(n_estimators = 200, max_depth=15, criterion="gini")  
clf.fit(xTrain, yTrain["IsBadBuy"].ravel())
y_pred = clf.predict_proba(xTest)[:,1]
roc_auc_score(yTest["IsBadBuy"], y_pred)
```
